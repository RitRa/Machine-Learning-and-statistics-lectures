# Two variables plots

adapted from https://matplotlib.org/tutorials/introductory/pyplot.html

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4])
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'rx')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'b-')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
x = np.linspace(0.0, 100.0, 101)
y = 2.0 * x

plt.plot(x, y, "k.")
```

```python
x = np.linspace(0.0, 100.0, 101)
y = -2.0 * x

plt.plot(x, y, "k.")
```

```python
x = np.linspace(0.0, 100.0, 101)
y1 = x * x 
y2 = x * x * x
y3 = x * x * x * x

plt.plot(x, y1, "r")
plt.plot(x, y2, "b")
plt.plot(x, y3, "g")
```
